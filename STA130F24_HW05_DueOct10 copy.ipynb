In a hypothesis test, researchers form hypotheses regarding the population parameters they are interested in, in this case, the population mean, $\mu$. The hypothesized population mean is denoted as $\mu_0$, where the null hypothesis states that the true population mean is equal to the hypothesized mean ($\mu=\mu_0$) and the alternative hypothesis states that the population mean is different from $\mu_0$. To test the hypotheses, researchers collect independent data $x_i\!$'s with a sample size n to form a sample and calculate the corresponding sample statistics, in this case, the sample mean, $\bar x$. The hypothesis test then results in a P-value that indicates the probability to obtain a sample with a sample statistics equal to or more extreme than $\bar x$, assuming the null hypothesis is true that the population mean is $\mu_0$. A small P-value indicates that the case is too extreme that such a sample is unlikely to be taken from a population with mean $\mu_0$, which results in rejecting the null hypothesis about the population mean $\mu$. Meanwhile, a large P-value implies that the case can happen by chance, resulting in a failure in rejecting the null hypothesis. Therefore, the hypothesis test uses sample statistics to help reseachers draw conclusions regarding the population parameters instead of sample statistics.

A p-value is the probability to obtain a sample with a sample statistics equal to or more extreme than the observed one, assuming the null hypothesis about the corresponding population parameter is true. In other words, it measures the probability to get the observed sample statistics from the presumed sampling distribution of the test statistics. As a result, a smaller p-value indicates that it is very unlikely for a value equal to or more extreme than the observed sample statistics to be obtained from the sampling distribution of the test statistics when the null hypothesis is true, creating a sense of 'ridiculousness' in the null hypothesis. For instance, the null hypothesis that a coin is fair would seem 'ridiculous' when the sample of 100 trials results in a sample proportion of 90% of front side landing upward, a case where the hypothesis test would result in a small p-value.

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

![Rodin's sculpture, "The Kiss"
](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d1/Rodin_-_Le_Baiser_06.jpg/409px-Rodin_-_Le_Baiser_06.jpg)


```python
import pandas as pd
from scipy import stats
import numpy as np

n_size = 124
p_null = 0.5
np.random.seed(1)

# Simulation of the null hypothesis
pop_distribution = stats.binom(n = n_size, p = p_null).rvs(size = 10000)/n_size
sample_stats = 0.645

# P-value
p_val = 2 * min(np.mean(pop_distribution <= sample_stats), np.mean(pop_distribution >= sample_stats))
print(f'P-value: {p_val}')

# Since 0.001 < p-value < 0.05, I have moderate evidence against the null hypothesis, and thus, 
# reject the null hypothesis.
```

# Tea Experiment Report

### 1. Introduction

In the $1920$s, statistician Ronald Fisher and his colleague Dr. Muriel Bristol had a dispute over whether Dr. Bristol could distinguish between a cup of tea with milk that involved milk being poured in first and another cup that had tea poured in first. In a simple experiment involving eight cups of tea with milk where four had milk added in first and four had tea added in first, Dr. Bristol successfully identified the pouring sequences of milk and tea of all the eight cups. Given this interesting anecdote, this experiment aims to investigate whether a  STA130 student is able to detect the pouring order of milk and tea for a cup of tea with milk on average. The null hypothesis of the experiment ($H_0$) is that an average STA130 student would not be able to detect the pouring order, and the alternative hypothesis ($H_a$) is that the STA130 students are able to detect the pouring order on average. 

### 2. Method
This experiment focuses on analyzing the pouring order detection ability among all STA130 students, which is the population of the study. To quantify the construct, this experiment would set the population parameter of interest to be the population proportion of students who are able to detect the pouring order of a cup of tea with milk. Denote this population proportion as $P$. Since the population proportion of people who can successfully name the pouring order entirely by chance in a population that has no ability to detect the pouring order would be expected to be fifty percent, set $P_0=0.5$, and the hypotheses of this experiment can be rewritten as: 
$$
H_0: P=P_0;  
H_a: P\neq P_0
$$

A hypothesis test is going to be conducted to test the hypotheses with significance level $\alpha=0.05$. The first step would be to collect data, for which a sample of eighty STA130 students has been randomly drawn from the population, each given a cup of tea with milk to determine the pouring order. The correctness of each students' answers form an independent sample of size $n=80$, forty-nine of which feature correct responds. Corresponding to the population parameter of interest, the sample statistics of interest in this experiment is the sample proportion of students who successfully named the pouring order, $\hat p$. Hence, the experiment results in the sample proportion $\hat p=\frac{49}{80}=0.6125$. 

Next, a simulation of the sampling distribution of the proportion is conducted with the code below under the null hypothesis that students are merely randomly guessing. It repeatedly selects random and independent samples from a binomial population with success probability of $0.5$ and sample size of $80$ for $10000$ times, and the sample proportions are calculated for each sample by dividing the number of success trials by the sample size. The simulated samples whose proportions differ from $P_0$ with absolute value of differences equal to or greater than the absolute difference between the observed proportion and $P_0$ represent similarly or more extreme occations than the observed sample in the hypothesized sampling distribution, and the proportion of these trials among the simulations serves as the p-value to test the hypothesis.

### 3. Result and Conclusion
The p-value, as shown below, is calculated to be $0.0465$, which is lower than the significance level $\alpha=0.05$ set at the beginning of the study. This indicates that the observed proportion of students correctly detecting the pouring order is rather unlikely to happen under the null hypothesis. This low possibility can also be observed in the graph below, where the area under the histogram of the ten thousand simulations to the right of the line representing the observed sample proportion is fairly small. Therefore, there is enough statistical evidence to reject the null hypothesis and conclude that the STA130 students are, on average, able to detect the pouring order of milk and tea for a cup of tea with milk.

```python
import numpy as np
from scipy import stats

np.random.seed(130) 
number_of_simulations = 10000 
n_size = 80
population_parameter = 0.5 # H0: Guessing (50% correct)

# Simulate the sampling distribution under the null hypothesis with the Binomial Distribution
simulated_correct_counts = stats.binom(n=n_size, p=population_parameter).rvs(size=number_of_simulations)
simulated_test_statistics = simulated_correct_counts / n_size

# Filter the extreme trials
observed_test_statistic = 49/80

SimTestStats_as_or_more_extreme_than_ObsTestStat = \
    abs(simulated_test_statistics - population_parameter) >= abs(observed_test_statistic - population_parameter)

# Calculate the p-value 
p_value = (SimTestStats_as_or_more_extreme_than_ObsTestStat).sum() / number_of_simulations
print("Number of Simulations: ", number_of_simulations, "\n\n",
      "Number of simulated test statistics (under HO)\n",
      'that are "as or more extreme" than the observed test statistic: ',
      SimTestStats_as_or_more_extreme_than_ObsTestStat.sum(), "\n\n",
      'p-value \n(= simulations "as or more extreme" / total simulations): ', p_value, sep="")



# A histogram of the sampling distribution of the 10000 trials and the corresponding position of the observed
# sample proportion

import plotly.express as px
import plotly.graph_objects as go
fig = px.histogram(simulated_test_statistics, nbins=20,
                   title="Histogram of Simulated Test Statistics with Observed Sample Proportion",
                   labels={'value': 'Simulated Test Statistics (Proportion Correct)'})

hist_data = np.histogram(simulated_test_statistics, bins=20)
y_max = hist_data[0].max()

# Add the vertical line for the observed test statistic using plotly.graph_objects
fig.add_trace(go.Scatter(
    x=[observed_test_statistic, observed_test_statistic],
    y=[0, y_max], mode="lines", line=dict(color="red", dash="dash"),
    name=f'Observed: {observed_test_statistic:.3f}'
))

# Update axis labels
fig.update_layout(xaxis_title="Simulated Test Statistics (Proportion Correct)", 
                  yaxis_title="Density")

# Show the plot
fig.show(renderer = 'png')
```

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

With this quite broad introductory context in mind, what does your favorite ChatBot thinks about the following statements? 
    
1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. So this therefore overlooks the possibility of sample size limitations, or Type II errors (which means a test incorrectly concludes that there is no effect or difference when, in fact, there is one). 
    
> Similarly, analagously, a "significant result" used to reject the null hypothsis could alternatively be a Type I error (which means a test actually incorrectly rejected a null hypothesis when it was actually true)... we're only providing a measure of evidence against the null hypothesis... but the evidence could still incorrectly suggest the wrong conclusion... it really depends on how strong the evidence is...
>
> - all of which is why just interpreting p-values using the table above is a good idea...

3. The p-values used for hypothesis testing are contructed upone the assumptions of the null hypotheses they correspond to; but, null hypotheses are actually often presented in simple forms that routinely hide a lot of information that is implicitly used to construct the p-values. For example, distributional assumptions about the population, estimated "plug-in" values that can used to simplify the problem calculations, and the reliance upon "random sampling", etc...<br><br>
           
4. Drawing overly broad conclusions, or making recommendations based on findings that reject the null hypothesis in a specific context is fraught with the problematic risks of overgeneralization errors. Further exacerbating this issue, null hypotheses are typically so called "point null hypotheses" which is meant to emphasize that they are mathematically increadibly sharply specific; whereas, alternative hypotheses are usually very unspecific. An alternative hypothesis that "the null hypothesis is false" doesn't say much... we should wonder, "how, specfically, is the null false?"
    
As an example really giving a demonstrating this, consider rejecting a null hypothesis that there is no correlation between rain and pizza's delivered. Such a decision doesn't specify what the actual hypothetical correlation might be. In fact, it doesn't even indicate if there are more or less pizzas delivered when it rains... 

> which, actually, shows very clearly why statistical inference using hypothesis testing is inferior to statistical inference based on confidence intervals...
> 
> - a confidence interval provides a range of plausible values of what the parameter in question might be; whereas, ...
> - trying to more clearly address what the plausible values of the parameter in question might be on the basis of hypothesis testing would require conducting further experiements to continously reject increasingly detailed hypothesies to narrow down what the alternative hypothesis might actually include... which would indeed be an utterly vapid misuse of the intended purpose of hypothesis testing entrprise...
